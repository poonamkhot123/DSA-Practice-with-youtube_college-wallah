**Sorting Algorithms**
 * In computer science,a sorting algorithm is an algorithm that puts elements of a list into an order.The most frequently used orders are numerical order and lexicographical order,and either ascending or descending.

**Sort an Array**
* **Bubble sort**
* given an array of integers nums,sort the array in ascending order and return it.
* you must solve the problem without using built in functions in o(nlog(n)) time complexity and with the smallest space complexity possible.
* input: nums = [5,2,3,1]
* output : [1,2,3,5]

```python
# Bubble Sort algorithm
# Space complexity :o(1)
# time complexity :worst case:O(n^2 )
#                  best case:omega(n)

def sortArray(nums):
  n = len(nums)
  for i in range(n):
    for j in range(n-i-1):
      if nums[j]>nums[j+1]:
       #swap
       temp = nums[j]
       nums[j] = nums[j+1]
       nums[j+1] = temp
  return nums

my_list = [3,4,5,7,1,2]
result = sortArray(my_list)
print(result)
```

```python
# Insertion sort algorithm
# space complexity :o(1)
# time complexity : worst case O(n^2)
#                   Best case omega(n)
def sortArray(nums):
  n = len(nums)
  for i in range(1,n):
    key = nums[i]
    j = i-1
    while j>=0 and nums[j]>key:
      nums[j+1]=nums[j]
      j-=1
    nums[j+1]= key
  return nums

my_list = [3,4,5,7,1,2]
result = sortArray(my_list)
print(result)
```

```python
# Selection sort algorithm
# space complexity :o(1)
# time complexity : worst case O(n^2)
#                   Best case omega(n^2)

def sortArray(nums):
  n = len(nums)
  for i in range(n):
    mn = nums[i]
    ind = i
    for j in range(i+1,n):
      if nums[j]<mn:
        mn = nums[j]
        ind = j
    temp = nums[i]
    nums[i] = nums[ind]
    nums[ind] = temp
  return nums

my_list = [3,4,5,7,1,2]
result = sortArray(my_list)
print(result)
```

```python
# Merge sort algorithm
# l : starting point
# r : ending point
def merge(nums,l,mid,r):
  a = []
  b = []

  for i in range(l,mid+1):
    a.append(nums[i])
  for i in range(mid+1,r+1):
    b.append(nums[i])

  i,j,k =0,0,l

  while i < len(a) and j < len(b):
    if a[i] < b[j]:
      nums[k] = a[i]
      i += 1
    else:
      nums[k] = b[j]
      j += 1
    k += 1

  while i < len(a):
    nums[k] = a[i]
    i += 1
    k += 1

  while j < len(b):
    nums[k] = b[j]
    j += 1
    k += 1


def mergeSort(nums,l,r):   # O(nlogn)
  # base case
  if l>=r:
    return

  # recursive case
  mid = (l+r)//2
  mergeSort(nums,l,mid) # Pass 'nums' here
  mergeSort(nums,mid+1,r) # Pass 'nums' here
  merge(nums,l,mid,r) # Merge the sorted subarrays

def sortArray(nums):
  mergeSort(nums,0,len(nums)-1) # Pass 'nums' here

  return nums
print(sortArray([1,3,4,54,57,68,34,56,90,76,45,34,456,34,234,5667]))
```

```python
# Quick Sort
# space complexity :o(1)
# time complexity : worst case O(n^2)
#                   Best case thita(nlog(n))

def partition(nums,l,r):
  key = nums[r]
  start = l

  for i in range(l,r): # Iterate up to r (exclusive)
    if nums[i]<=key: # Condition should be <= key for elements smaller than or equal to pivot
      temp = nums[i]
      nums[i] = nums[start]
      nums[start]=temp
      start+=1

  # Swap pivot with element at 'start'
  temp = nums[r]
  nums[r] = nums[start]
  nums[start] = temp

  return start # Return the pivot's final position

def quicksort(nums,l,r):
  # base case
  if l >=r:
    return
  p = partition(nums,l,r)
  quicksort(nums,l,p-1)
  quicksort(nums,p+1,r)

def sortArray(nums):
  n = len(nums)
  quicksort(nums, 0, n-1) # Pass the initial left (0) and right (n-1) indices
  return nums

print(sortArray([1,3,4,54,57,68,34,56,90,76,45,34,456,34,234,5667]))
```

```python
# Counting Sort
# we use when limited numbers are their
def sortArray(nums):
  n= len(nums)
  mx = max(nums)
  freq = [0]*(mx+1)

  for i in range(0,mx+1):
    while freq[i]>0:
      nums.append(i)
      freq[i]-=1
  return nums
```

**Merge Sorted Array**
* you are given two integer arrays nums1 and nums2 ,sorted in non-decreasing order,and two integers m and n ,representing the number of elements in nums1 and nums2 respectively.
* Merge nums1 and nums2 into a single array sorted in non- decreasing order.
 * The final sorted array should not be returned by the function , but instead be sorted inside the array nums1. To accomodate this,nums1 has a length of m+n,where the first m elements denote the elements that should be merged , and the last elements are set to 0 and should be ignored. nums2 has a length of n.
 * input : nums1 =[1,2,3,0,0,0] ,m =3,
 *         nums2 = [2,5,6] , n= 3
 *         output : [1,2,2,3,5,6]
 * Explanation: The arrays we are merging are [ 1,2,3] and [2,5,6].
 * the result of the merge is [1,2,2,3,5,6] with the underlined elements coming from nums1.

```python
def merge(nums1,m,n):
  i = m-1
  j = n-1
  k = m+n-1

  while j>=0:
    if i<0 or nums2[j]>nums1[i]:
      nums1[k]=nums2[j]
      k-=1
      j-=1
    else:
      nums1[k]=nums1[i]
      k-=1
      i-=1
  return nums1 # The problem statement says to sort in place, but returning the modified list is also common practice.

nums1 =[1,2,3,0,0,0]
nums2 = [2,5,6]
m = 3
n = 3
result = merge(nums1, m, n) # Pass m and n here
print(result)
```

```python
def eraseOverlapIntervals(intervals):
  intervals.sort(key = lambda x: x[1])
  n = len(intervals)

  prev = 0
  count = 1

  for i in range(1,n):
    if intervals[i][0] >= intervals[prev][1]:
      count+=1
      prev = i
  return n - count

intervals = [[1,2],[2,3],[3,4],[1,3]]
result = eraseOverlapIntervals(intervals)
print(result)
```

**Sort Array by Parity**
* given an integer array nums ,move all the even integers at the beginning of the array followed by all the odd integers.

* input : nums= [3,1,2,4]
* output : [2,4,3,1]
* Explanation : The output [4,2,3,1] , [2,4,1,3] , and [4,2,1,3] would also be accepted.

```python
def sortarraybyparity(nums):
  start = 0
  for i in range(len(nums)):
    if nums[i]%2==0:
      temp = nums[i]
      nums[i]=nums[start]
      nums[start]=temp
      start+=1
  return nums

nums = [3,4,5,6,7,8,6,3,1,2]
result = sortarraybyparity(nums)
print(result)
```

```python
# Binary Search
# Search 2D matrix
def searchmatrix(matrix,target):
  rows = len(matrix)
  cols = len(matrix[0])

  l = 0
  r = rows * cols - 1

  while l<=r:
    mid = (l+r)//2

    i = mid//cols
    j = mid % cols

    if matrix[i][j] == target:
      return True
    elif matrix[i][j] > target:
      # left
      r = mid -1
    else:
      # right
      l = mid+1
    return False




```
