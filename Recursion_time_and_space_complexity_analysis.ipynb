```python
# Recursion
def printNumbers(i,n):
  # base case
  if i >n:
    return

  # recursive case
  print(i,end = " ")
  printNumbers(i+1,n)

printNumbers(1,5)
```

```python
# recursion example: factorial of a number.
def fact(n):
  if n==0:
    return 1
  return n*fact(n-1)
print(fact(4))
```

```python
def fun(n):
  if n==0:
    return
  fun(n-1)

  print(n,end=" ")

fun(5)
```

```python
# fibonacci number
the fibonacci numbers ,commomly denoted F(n) form a sequence,
called the Fibonacci sequence,such that each number is the
sum of the two preceding ones,starting from 0  and 1.
```

```python
def fib(n):
  # base case
  if n ==0 or n==1:
    return n

  # recursive case
  return fib(n-1)+fib(n-2)

fib(7)
```

```python
# Power of Two
Given an interger n ,return true if it is a power of two.Otherwise ,
return false.
An integer n is a power of two,if there exists an integer x such that
n==2^x.

input: n =16
output:true
Explanation: 2^4 = 16
```

```python
# by using while loop.
def isPowerOfTwo(n):
  while n%2==0:
    n//=2

  return n==1
```

```python
# by using recursion.
def isPowerofTwo(n):
  # base case
  if(n<=0):
   return True
  if n==1:
    return True
  if n%2!=0:
    return False
  #recursive case
  return isPowerofTwo(n//2)
isPowerofTwo(4)
```

```python
# GCD of two numbers - Euclidean algorithm.
The Euclidean algorithm is a method for efficiently calculating the greatest
common divisor (GCD) of two integers.
```

```python
def gcd(a,b):
  if b==0:
    return a
  return gcd(b,a%b)
print(gcd(15,50))
```

```python
# find LCM with the help of GCD.
def gcd(a,b):
  if b==0:
    return a
  return gcd(b,a%b)

def lcm(a,b):
  return a*b/gcd(a,b)

print(gcd(15,50))
print(lcm(15,50))
```
