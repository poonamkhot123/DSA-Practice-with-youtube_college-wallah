```python
# Runnig sum of n d array.
def runningSum(nums):
  n = len(nums)

  ans = []
  ans.append(nums[0])

  for i in range(1,n):
    x = ans[i-1] + nums[i]
    ans.append(x)

  return ans
s = [1,2,3,4,5]
print(runningSum(s))
```

```python
# Remove Duplicates from sorted array
def removeDuplicates(nums):
  n = len(nums)

  start = 0
  for i in range(1,n):
    # unique element
    if nums[i] !=nums[start]:
      start +=1
      nums[start] = nums[i]
  return start+1
```
