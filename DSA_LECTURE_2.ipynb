**TIME COMPLEXITY**
* Time cmplexity in computer science describes how the execution time of an algorithm grows as the input size increases.
* It eassentially quantifies the number of operations an algorithm performs as a function of the input data size.

```python
n = 10
x = 3
for i in range(n):
  if i == x:
    break
  print(i,end=" ")
```

* **Best Case(minimum number of operations)** -> 1 operation......omega(1)
* **worst Case(maximum number of operations)** -> n operations......bigO(n)
* **average case** -> n/2 operations.....thita(n/2)

# **Introduction to Recursion**
* The process in which a function calls itself directly or indirectly is called recursion.

* and the corresponding function is called a recursive function.

```python
# print 1 to n using Recursion.
def printNumbers(i,n):
  # base case
  if i>n:
    return
  # recursive case
  print(i,end=" ")
  print(i+1,n)
printNumbers(1,5)
```

```python
def fun(n):
  if n==0:
    return
  print(n,end=" ")
  fun(n-1)
fun(3)
```

# **Recursive Tree**

**Fibonacci Number**
* the fibonacci numbers,commonly denoted F(n) form a sequence,called the Fibonacci sequence,such that each number is the sum of the two proceding ones,starting from 0 to 1.That is,

* F(0) = 0
* F(1) = 1
* F(n) = F(n-1)+F(n-2), for n>1

```python
def fib(self,n:int):
  #base case
  if n== 0 or n==1:
    return n

  #recursive case
  return self.fib(n-1) + self.fib(n-2)

```

```python
# Example
# Given an integer n ,returns true if it is a power of two.
# Otherwise ,return false.
# An integer n is a power of two,if there exits an integer x such that n == 2^x
# Example:1,2,4,8,16,32,64,128

class solution:
  def isPoweroftwo(self,n):
    # base case
    if(n<=0):
       return False
    if n==1:
       return True
    if n%2!=0:
      return False

    # Recursive case
    return self.isPoweroftwo(n//2)

```

# **GCD of Two numbers- Euclidean algorithm**
the euclidean algorithm is a method for efficiently calculating the greatest common divisor (GCD) of two integers.

```python
def gcd(a,b):
  if b==0:
    return a
  return gcd(b,a%b)
print(gcd(15,50))
```

```python
# to find the LCM WE KNOW THAT [a*b = gcd(a,b) * lcm(a,b)]
def gcd(a,b):
  if b==0:
    return a
  return gcd(b,a%b)

def lcm(a,b):
  return a*b/gcd(a,b)

print(gcd(15,50))
print(lcm(15,50))
```
